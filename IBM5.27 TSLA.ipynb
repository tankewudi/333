```python
# Import required libraries
import yfinance as yf
import matplotlib.pyplot as plt
import pandas as pd
import requests
from bs4 import BeautifulSoup

# Problem 1: Extract Tesla stock data using yfinance (2 points)
def problem1():
    print("Problem 1: Tesla Stock Data")
    tsla_data = yf.download('TSLA', start='2023-01-01', end='2024-01-01')
    tsla_data['Date'] = tsla_data.index
    tsla_data = tsla_data[['Date', 'Close']]
    
    plt.figure(figsize=(12, 6))
    plt.plot(tsla_data['Date'], tsla_data['Close'], color='red')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title('Tesla Closing Prices (2023)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return tsla_data

# Problem 2: Web scrape Tesla revenue data (1 point)
def problem2():
    print("Problem 2: Tesla Revenue Data")
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    html_data = requests.get(url).text
    soup = BeautifulSoup(html_data, 'html.parser')
    
    tables = soup.find_all('table')
    for table in tables:
        if 'Tesla Quarterly Revenue' in str(table):
            revenue_data = pd.read_html(str(table))[0]
            revenue_data.columns = ['Date', 'Revenue']
            revenue_data['Revenue'] = revenue_data['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
            break
    
    print(revenue_data.head())
    return revenue_data

# Problem 3: Extract GameStop stock data using yfinance (2 points)
def problem3():
    print("Problem 3: GameStop Stock Data")
    gme_data = yf.download('GME', start='2023-01-01', end='2024-01-01')
    gme_data['Date'] = gme_data.index
    gme_data = gme_data[['Date', 'Close']]
    
    plt.figure(figsize=(12, 6))
    plt.plot(gme_data['Date'], gme_data['Close'], color='green')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title('GameStop Closing Prices (2023)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return gme_data

# Problem 4: Web scrape GameStop revenue data (1 point)
def problem4():
    print("Problem 4: GameStop Revenue Data")
    url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
    html_data = requests.get(url).text
    soup = BeautifulSoup(html_data, 'html.parser')
    
    tables = soup.find_all('table')
    for table in tables:
        if 'GameStop Quarterly Revenue' in str(table):
            revenue_data = pd.read_html(str(table))[0]
            revenue_data.columns = ['Date', 'Revenue']
            revenue_data['Revenue'] = revenue_data['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
            break
    
    print(revenue_data.head())
    return revenue_data

# Problem 5: Tesla stock and revenue dashboard (2 points)
def problem5(tsla_stock, tsla_revenue):
    print("Problem 5: Tesla Dashboard")
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Stock price plot
    ax1.plot(tsla_stock['Date'], tsla_stock['Close'], color='red')
    ax1.set_title('Tesla Stock Price (2023)')
    ax1.set_ylabel('Price (USD)')
    ax1.grid(True)
    
    # Revenue bar chart
    ax2.bar(tsla_revenue['Date'], tsla_revenue['Revenue'], color='orange')
    ax2.set_title('Tesla Quarterly Revenue')
    ax2.set_ylabel('Revenue (Millions USD)')
    ax2.grid(True)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Problem 6: GameStop stock and revenue dashboard (2 points)
def problem6(gme_stock, gme_revenue):
    print("Problem 6: GameStop Dashboard")
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Stock price plot
    ax1.plot(gme_stock['Date'], gme_stock['Close'], color='green')
    ax1.set_title('GameStop Stock Price (2023)')
    ax1.set_ylabel('Price (USD)')
    ax1.grid(True)
    
    # Revenue bar chart
    ax2.bar(gme_revenue['Date'], gme_revenue['Revenue'], color='blue')
    ax2.set_title('GameStop Quarterly Revenue')
    ax2.set_ylabel('Revenue (Millions USD)')
    ax2.grid(True)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Execute all problems
if __name__ == "__main__":
    # Run problems 1-4 to get data
    tsla_stock = problem1()
    tsla_revenue = problem2()
    gme_stock = problem3()
    gme_revenue = problem4()
    
    # Run dashboards (problems 5-6)
    problem5(tsla_stock, tsla_revenue)
    problem6(gme_stock, gme_revenue)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import requests
from bs4 import BeautifulSoup

# Problem 1: Load Tesla stock data from TSLA.csv (2 points)
def problem1():
    print("Problem 1: Tesla Stock Data (from TSLA.csv)")
    tsla_data = pd.read_csv('TSLA.csv', parse_dates=['Date'])  # 确保 CSV 文件中有 'Date' 列
    tsla_data = tsla_data[['Date', 'Close']]  # 仅保留日期和收盘价
    
    plt.figure(figsize=(12, 6))
    plt.plot(tsla_data['Date'], tsla_data['Close'], color='red')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title('Tesla Closing Prices (from TSLA.csv)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return tsla_data

def problem2():
    print("Problem 2: Tesla Revenue Data")
    url = "https://www.macrotrends.net/stocks/charts/TSLA/tesla/revenue"
    
    try:
        html_data = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).text
        soup = BeautifulSoup(html_data, 'html.parser')
        
        tables = soup.find_all('table')
        revenue_data = None  # 初始化变量
        
        for table in tables:
            if 'Tesla Quarterly Revenue' in str(table):
                revenue_data = pd.read_html(str(table))[0]
                revenue_data.columns = ['Date', 'Revenue']
                revenue_data['Revenue'] = revenue_data['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
                break
        
        if revenue_data is None:
            print("Warning: Tesla revenue table not found!")
            return pd.DataFrame(columns=['Date', 'Revenue'])  # 返回空 DataFrame 避免报错
        
        print(revenue_data.head())
        return revenue_data
    
    except Exception as e:
        print(f"Error fetching Tesla revenue data: {e}")
        return pd.DataFrame(columns=['Date', 'Revenue'])  # 出错时返回空 DataFrame

# Problem 3: Extract GameStop stock data using yfinance (2 points)
def problem3():
    print("Problem 3: GameStop Stock Data")
    gme_data = yf.download('GME', start='2023-01-01', end='2024-01-01')
    gme_data['Date'] = gme_data.index
    gme_data = gme_data[['Date', 'Close']]
    
    plt.figure(figsize=(12, 6))
    plt.plot(gme_data['Date'], gme_data['Close'], color='green')
    plt.xlabel('Date')
    plt.ylabel('Closing Price (USD)')
    plt.title('GameStop Closing Prices (2023)')
    plt.xticks(rotation=45)
    plt.grid(True)
    plt.tight_layout()
    plt.show()
    
    return gme_data

def problem4():
    print("Problem 4: GameStop Revenue Data")
    url = "https://www.macrotrends.net/stocks/charts/GME/gamestop/revenue"
    
    try:
        html_data = requests.get(url, headers={'User-Agent': 'Mozilla/5.0'}).text
        soup = BeautifulSoup(html_data, 'html.parser')
        
        tables = soup.find_all('table')
        revenue_data = None  # 初始化变量
        
        for table in tables:
            if 'GameStop Quarterly Revenue' in str(table):
                revenue_data = pd.read_html(str(table))[0]
                revenue_data.columns = ['Date', 'Revenue']
                revenue_data['Revenue'] = revenue_data['Revenue'].str.replace('$', '').str.replace(',', '').astype(float)
                break
        
        if revenue_data is None:
            print("Warning: GameStop revenue table not found!")
            return pd.DataFrame(columns=['Date', 'Revenue'])  # 返回空 DataFrame 避免报错
        
        print(revenue_data.head())
        return revenue_data
    
    except Exception as e:
        print(f"Error fetching GameStop revenue data: {e}")
        return pd.DataFrame(columns=['Date', 'Revenue'])  # 出错时返回空 DataFrame


# Problem 5: Tesla stock and revenue dashboard (2 points)
def problem5(tsla_stock, tsla_revenue):
    print("Problem 5: Tesla Dashboard")
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Stock price plot
    ax1.plot(tsla_stock['Date'], tsla_stock['Close'], color='red')
    ax1.set_title('Tesla Stock Price (from TSLA.csv)')
    ax1.set_ylabel('Price (USD)')
    ax1.grid(True)
    
    # Revenue bar chart
    ax2.bar(tsla_revenue['Date'], tsla_revenue['Revenue'], color='orange')
    ax2.set_title('Tesla Quarterly Revenue')
    ax2.set_ylabel('Revenue (Millions USD)')
    ax2.grid(True)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Problem 6: GameStop stock and revenue dashboard (2 points)
def problem6(gme_stock, gme_revenue):
    print("Problem 6: GameStop Dashboard")
    
    fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(12, 10))
    
    # Stock price plot
    ax1.plot(gme_stock['Date'], gme_stock['Close'], color='green')
    ax1.set_title('GameStop Stock Price (2023)')
    ax1.set_ylabel('Price (USD)')
    ax1.grid(True)
    
    # Revenue bar chart
    ax2.bar(gme_revenue['Date'], gme_revenue['Revenue'], color='blue')
    ax2.set_title('GameStop Quarterly Revenue')
    ax2.set_ylabel('Revenue (Millions USD)')
    ax2.grid(True)
    
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()

# Execute all problems
if __name__ == "__main__":
    # Run problems 1-4 to get data
    tsla_stock = problem1()  # 从 TSLA.csv 加载数据
    tsla_revenue = problem2()  # 爬取特斯拉收入数据
    gme_stock = problem3()  # 获取 GameStop 股票数据
    gme_revenue = problem4()  # 爬取 GameStop 收入数据
    
    # Run dashboards (problems 5-6)
    problem5(tsla_stock, tsla_revenue)
    problem6(gme_stock, gme_revenue)
```
